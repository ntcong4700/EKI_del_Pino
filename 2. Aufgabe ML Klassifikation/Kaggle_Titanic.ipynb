```python
import pandas as pd
```

```python
train_data = pd.read_csv("./data/train.csv")
test_data = pd.read_csv("./data/test.csv")
```

```python
#Wie wir sehen in describe(), wir haben nur 714 / 891 Age , 889 / 891 Embarked 
# => Fehlende Werte mit dem Mittelwert aller Werte auffuhllen
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imp = SimpleImputer(missing_values=np.nan, strategy='median')
train_data['Age'] = imp.fit_transform(train_data[['Age']])
```

```python
train_data['Embarked'] = train_data['Embarked'].fillna('C')
test_data.head()
```

```python
test_data.describe()
```

```python
test_data['Age'] = imp.fit_transform(test_data[['Age']])
test_data['Embarked'] = test_data['Embarked'].fillna('C')
```

```python
# Wir umwandeln Sex zu number
# Female = 1
# Male = 0
train_data['Sex'] = train_data['Sex'].replace(['female', 'male'], [0, 1])
test_data['Sex'] = test_data['Sex'].replace(['female', 'male'], [0, 1])
```

```python
# C = Cherbourg
# Q = Queenstown
# S = Southampton
```

```python
train_data['Embarked'] = train_data['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2])
test_data['Embarked'] = test_data['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2])
```

```python
features = ["Pclass", "Sex", "Age", "SibSp", "Fare", "Embarked"]
x = train_data[features]
y = train_data["Survived"]
X_test = test_data[features]
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x, y)
```

```python
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
```

```python
output.to_csv('submission.csv', index=False)
```
